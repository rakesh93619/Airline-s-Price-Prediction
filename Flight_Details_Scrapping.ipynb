```python
## importing the libraries:
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
import time

import re

## importing require exceptions:-
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException,WebDriverException 

import warnings
warnings.filterwarnings('ignore')
```

## 1. Yatra.com

```python
## fetching the  website:
driver.get('https://www.yatra.com/')
time.sleep(3)

## Clicking on search flight button
driver.find_element_by_xpath('//input[@value="Search Flights"]').click()

```

### Scrapping All The details

```python
Flight_name=[]
Departure_time=[]
Arrival_time=[]
Source=[]
Destination=[]
Duration=[]
Total_stops=[]
Route=[]
Price=[]

try:
    tags=driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')
    for i in tags:
        Flight_name.append(i.text)
except NoSuchElementException:
        Flight_name.append('-')

try:
    tags=driver.find_elements_by_xpath('//div[@autom="departureTimeLabel"]')
    for i in tags:
        Departure_time.append(i.text)
except NoSuchElementException:
        Departure_time.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//p[@autom="arrivalTimeLabel"]')
    for i in tags:
        Arrival_time.append(i.text)
except NoSuchElementException:
        Arrival_time.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
    count=1
    for i in tags:
        if count%2!=0:
            Source.append(i.text)
        count+=1
except NoSuchElementException:
        Source.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
    count=1
    for i in tags:
        if count%2==0:
            Destination.append(i.text)
        count+=1
except NoSuchElementException:
        Destination.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//p[@autom="durationLabel"]')
    for i in tags:
        Duration.append(i.text)
except NoSuchElementException:
        Duration.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
    for i in tags:
        Total_stops.append(i.text)
except NoSuchElementException:
        Total_stops.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
    for i in tags:
        Price.append(i.text)
except NoSuchElementException:
        Price.append('-')
```

```python
print(len(Flight_name),len(Departure_time),len(Arrival_time),len(Source),len(Destination),len(Duration),len(Total_stops),
      len(Price))
```

## Creating A DataFrame

```python
date=['30/05/2022']
website=['Yatra.com']
```

```python
## Addting other feature as well for Mumbai location:
df=pd.DataFrame()
df['Flight_name']=Flight_name
df['Departure_time']=Departure_time
df['Arrival_time']=Arrival_time
df['Source']=Source
df['Destination']=Destination
df['Duration']=Duration
df['Total_stops']=Total_stops
df['Date_of_journey']=date*177
df['Website_name']=website*177
df['Price']=Price
```

```python
try:
    tags=driver.find_elements_by_xpath('//span[@class="txt-r4 ng-binding"]')
    for i in tags:
        Flight_name.append(i.text)
except NoSuchElementException:
        Flight_name.append('-')

try:
    tags=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-4 top5 "]')
    for i in tags:
        Departure_time.append(i.text.split('\n')[0])
    
except NoSuchElementException:
        Departure_time.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-3 top5"]')
    for i in tags:
        Arrival_time.append(i.text.split('\n')[0])
except NoSuchElementException:
        Arrival_time.append('-')
        
        
try:
    tags=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-5 non-st"]')
    for i in tags:
        Duration.append(i.text.split('\n')[0])
except NoSuchElementException:
        Duration.append('-')
try:
    tags=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-4 top5 "]')
    for i in tags:
        Source.append(i.text.split('\n')[1])
except NoSuchElementException:
        Source.append('-')
try:
    tags=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-3 top5"]')
    for i in tags:
        Destination.append(i.text.split('\n')[1])
except NoSuchElementException:
        Destination.append('-')
try:
    tags=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-5 non-st"]//span[2]')
    for i in tags:
        Total_stops.append(i.text)
except NoSuchElementException:
    Total_stops.append('-')
        
try:
    tags=driver.find_elements_by_xpath('//div[@class="col-md-2 col-sm-2 col-xs-5 mr5 cle"]')
    for i in tags:
        Price.append(i.text.split('\n')[0])
except NoSuchElementException:
        Price.append('-')
```

```python
print(len(Flight_name),len(Departure_time),len(Arrival_time),len(Source),len(Destination),len(Duration),len(Total_stops),
      len(Price))
```
